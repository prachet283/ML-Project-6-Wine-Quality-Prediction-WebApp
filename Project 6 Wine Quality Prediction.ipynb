Importing dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

Data Collection

```python
# loading the dataset to a Pandas Dataframe
wine_dataset = pd.read_csv('/content/ML-Project 6 winequality-red.csv')
```

```python
wine_dataset['quality'].value_counts()
```

Data Analysis & Visualization

```python
# statistical measures of data
wine_dataset.describe()
```

Data Preprocessing

```python
# seperate the data and label
X = wine_dataset.drop('quality', axis=1)
X.head()
```

Label Binarization

```python
Y = wine_dataset['quality'].apply(lambda y_value: 1 if y_value>=6 else 0)
```

```python
Y
```

Training & Test Split

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(Y.shape , Y_train.shape , Y_test.shape)
```

Model Training:
Random Forest Classifier

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation"

```python
#accuracy on train data
X_train_predication = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_predication , Y_train)
```

```python
print('Train Accuracy = ' , train_data_accuracy)
```

```python
#accuracy on test data
X_test_predication = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predication , Y_test)
```

```python
print('Test Accuracy = ' , test_data_accuracy)
```

Building a Predictive System

```python
input_data = (8.2,0.4,0.44,2.8,0.08900000000000001,11.0,43.0,0.9975,3.53,0.61,10.5)
#changing the input data in numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

if(prediction[0]==1):
  print("Good Quality Wine")
else:
  print("Bad Quality Wine")
```
